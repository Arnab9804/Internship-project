# Traning & Testing the data

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error,mean_squared_error
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.linear_model import LinearRegression
```

```python
maxacc=0
maxre=0
for i in range(1,200):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30,random_state=i)
    lr=LinearRegression()
    lr.fit(x_train,y_train)
    predict_lr=lr.predict(x_test)
    acc=r2_score(y_test,predict_lr)
    if acc>maxacc:
        maxacc=acc
        maxre=i
print('Maximum r2 score is:',maxacc,'Best Random Estate:',maxre)
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x=predict_lr,y=y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.ensemble import ExtraTreesRegressor
from sklearn.svm import SVR
from sklearn.linear_model import SGDRegressor

from sklearn.model_selection import cross_val_score
```

```python
def regressionModel(model):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30,random_state=72)
    model.fit (x_train,y_train)
    print('Model traning Score:',model.score(x_train,y_train))
    pred_model=model.predict (x_test)
    print ('Best r2 score:',r2_score(pred_model,y_test))
    print ('Mean absolute Error Score:',mean_absolute_error(pred_model,y_test))
    print ('Mean squared error Score:', mean_squared_error(pred_model,y_test))
    print ('Root Mean squared error:',np.sqrt(mean_squared_error(pred_model,y_test)))
    
    
    score=cross_val_score(model,x,y,cv=7)
    print('CV Score:', np.mean(score))
    print ('Difference between r2 score and cv score:', r2_score(pred_model,y_test) - np.mean(score))
    
    
```

```python
model=RandomForestRegressor()                                                                         # R2_Score: 0.92
regressionModel(model)                                                                                # CV Score: 0.90
```

```python
model=LinearRegression()                                                                              # R2_Score: 0.96
regressionModel(model)                                                                                # CV Score: 0.90
```

```python
model=DecisionTreeRegressor()                                                                         # R2_Score: 0.85
regressionModel(model)                                                                                # CV Score: 0.75
```

```python
model=KNeighborsRegressor()                                                                         # R2_Score: 0.96
regressionModel(model)                                                                              # CV Score: 0.88
```

```python
model=Ridge()                                                                                       # R2_Score: 0.96
regressionModel(model)                                                                              # CV Score: 0.90
```

```python
model=Lasso()                                                                                       # R2_Score: -12.67
regressionModel(model)                                                                              # CV Score: 0.33
```

```python
model=ExtraTreesRegressor()                                                                         # R2_Score: 0.95
regressionModel(model)                                                                              # CV Score: 0.92
```

```python
model=SVR()                                                                                         # R2_Score: 0.97
regressionModel(model)                                                                              # CV Score: 0.91
```

```python
model=SGDRegressor()                                                                                # R2_Score: 0.5
regressionModel(model)                                                                              # CV Score: 0.74
```

After evaluating various regression models based on the difference in R2 scores and cross-validation scores, SVRr is identified as the most suitable model for the given dataset.

# Data Modeling

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix
from sklearn.linear_model import LogisticRegression
```

```python


maxAcc=0
maxRE=0
for i in range (1,200):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30,random_state=i)
    lg=LogisticRegression()
    lg.fit (x_train,y_train)
    pred_lg=lg.predict(x_test)
    acc=accuracy_score(pred_lg,y_test)
    if acc>maxAcc:
        maxAcc=acc
        maxRE=i

print ("Best accuracy score:",maxAcc,"Best random Estate:",maxRE)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.svm import SVC
from sklearn.ensemble import GradientBoostingClassifier,AdaBoostClassifier,BaggingClassifier
from sklearn.tree import DecisionTreeClassifier

from sklearn.model_selection import cross_val_score
```

```python
def classification_Model(model):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30,random_state=114)
    model.fit (x_train, y_train)
    pred_model=model.predict(x_test)
    print ("Best Model Score:",model.score(x_train,y_train))

    print ('Best Accuracy score:',accuracy_score(pred_model,y_test))
    
    print ('Best Classification Report:', classification_report(pred_model,y_test))
    
    print ('Best confusion Matrix:', confusion_matrix(pred_model,y_test))
    
    print ('\n')
    
    score=cross_val_score(model,x,y,cv=7)
    print ("Best CV Score:",np.mean(score))

    
```

```python
model=LogisticRegression()
classification_Model(model)
```

```python
model=RandomForestClassifier()
classification_Model(model)
```

```python
model=ExtraTreesClassifier()
classification_Model(model)
```

```python
model=GaussianNB()
classification_Model(model)
```

```python
model=SVC()
classification_Model(model)
```

```python
model=GradientBoostingClassifier()
classification_Model(model)
```

```python
model=AdaBoostClassifier()
classification_Model(model)
```

```python
model=BaggingClassifier()
classification_Model(model)
```

```python
model=DecisionTreeClassifier()
classification_Model(model)
```

##### Observation
After evaluating various regression models based on Accuracy score and cross-validation scores, ExtraTreesClassifier is identified as the most suitable model for the given dataset.

#### BUILD MODEL

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
from sklearn.linear_model import LogisticRegression
```

```python
maxAC=0
maxRE=0
for i in range (1,200):
    x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=.20,random_state=i)
    lg=LogisticRegression()
    lg.fit (x_train,y_train)
    pred_lg=lg.predict(x_test)
    acc=accuracy_score(pred_lg,y_test)
    if acc>maxAC:
        maxAC=acc
        maxRE=i
print('Best Accuracy Score:', maxAC,'Best RandomEstate:', maxRE)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.svm import SVC
from sklearn.ensemble import GradientBoostingClassifier,AdaBoostClassifier,BaggingClassifier
from sklearn.tree import DecisionTreeClassifier

from sklearn.model_selection import cross_val_score
```

```python
def classmod(model):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=49)
    model.fit (x_train,y_train)
    pred_model=model.predict (x_test)
    print ('Model traning Score:',model.score (x_train,y_train))
    print ('Best Accuracy Score:',accuracy_score(pred_model,y_test))
    print('Confusion matrix value:', confusion_matrix(pred_model,y_test))
    print('\n')
    
    score=cross_val_score(model,x,y,cv=7)
    print ('Best CV Score:',np.mean (score))
    
    
```

```python
model=LogisticRegression()
classmod(model)
```

```python
model=DecisionTreeClassifier()
classmod(model)
```

```python
model=ExtraTreesClassifier()
classmod(model)
```

```python
model=RandomForestClassifier()
classmod(model)
```

```python
model=GradientBoostingClassifier()
classmod(model)
```

```python
model=GaussianNB()
classmod(model)
```

##### OBSERVATION

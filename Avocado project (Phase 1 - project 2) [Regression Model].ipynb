# Model Build

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error,mean_squared_error
from sklearn.linear_model import LinearRegression
```

```python
maxACC=0
maxRE=0
for i in range (1,200):
    x_train, x_test,y_train,y_test=train_test_split(x,y,test_size=.30,random_state=i)
    lr=LinearRegression()
    lr.fit (x_train,y_train)
    predict_lr=lr.predict(x_test)
    acc=r2_score(y_test,predict_lr)
    if acc>maxACC:
        maxACC=acc
        maxRE=i
print('Maximum r2 score is:',maxACC,'On Random Estate:',maxRE)
print ("Model build score:",lr.score(x_train,y_train))
        
```

##### Try with Different Model

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.ensemble import ExtraTreesRegressor
from sklearn.svm import SVR
from sklearn.linear_model import SGDRegressor

from sklearn.model_selection import cross_val_score
```

```python
def regression_Model(model):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30,random_state=6)
    model.fit (x_train, y_train)
    pred_model=model.predict(x_test)
    print ("Best Model Score:",model.score(x_train,y_train))

    print ('Best R2 score:',r2_score(pred_model,y_test))
    
    print ('Mean absolute Error Score:', mean_absolute_error(pred_model,y_test))
    
    print ('Mean squared Error Score:', mean_squared_error(pred_model,y_test))
    
    print ('Root Mean squared error:',np.sqrt(mean_squared_error(pred_model,y_test)))
    print ('\n')
    
    score=cross_val_score(model,x,y,cv=7)
    print ("Best CV Score:",np.mean(score))
    print ("Difference between r2 score and cv score:", r2_score(pred_model,y_test) - np.mean(score))
    
```

```python
model=LinearRegression()
regression_Model(model)
```

```python
model=DecisionTreeRegressor()
regression_Model(model)
```

```python
model=KNeighborsRegressor()
regression_Model(model)
```

```python
model=Ridge()
regression_Model(model)
```

```python
model=Lasso()
regression_Model(model)
```

```python
model=ExtraTreesRegressor()
regression_Model(model)
```

```python
model=SVR()
regression_Model(model)
```

```python
model=SGDRegressor()
regression_Model(model)
```

##### Observation
After evaluating various regression models based on the difference in R2 scores and cross-validation scores, Ridge is identified as the most suitable model for the given dataset.

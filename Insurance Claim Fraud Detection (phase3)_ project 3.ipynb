#### VIF CHECKING

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
def vif_cal():
    vif=pd.DataFrame()
    vif['VIF feature']=[variance_inflation_factor (x.values,i) for i in range (len (x.columns))]
    vif['Factors']=x.columns
    return (vif)
```

#### MODEL BUILD UP

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix
from sklearn.linear_model import LogisticRegression
```

```python
maxAcc=0
maxRE=0
for i in range (1,200):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30,random_state=i)
    lg=LogisticRegression()
    lg.fit (x_train,y_train)
    pred_lg=lg.predict(x_test)
    acc=accuracy_score(pred_lg,y_test)
    if acc>maxAcc:
        maxAcc=acc
        maxRE=i

print ("Best accuracy score:",maxAcc,"Best random Estate:",maxRE)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.svm import SVC
from sklearn.ensemble import GradientBoostingClassifier,AdaBoostClassifier,BaggingClassifier
from sklearn.tree import DecisionTreeClassifier

from sklearn.model_selection import cross_val_score
from sklearn import metrics
```

```python
model_class={}
def classify_model(model):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=42)
    model.fit (x_train,y_train)
    pred_model=model.predict (x_test)
    print ('Best Trening Model Score:',model.score(x_train,y_train))
    print ('Best Accuracy Score:',accuracy_score(pred_model,y_test))
    print ('\n')
    
    score=cross_val_score(model,x,y,cv=9)
    print('Best CV Score:',np.mean(score))
    model_class[model]=round (np.mean(score)*100,2)
    
    fpr,tpr,thresold=metrics.roc_curve(y_test,pred_model)
    roc_auc=metrics.auc(fpr,tpr)
    display=metrics.RocCurveDisplay(fpr=fpr,tpr=tpr,roc_auc=roc_auc,estimator_name=model)
    display.plot()
    
    
```

```python
model=LogisticRegression()
classify_model(model)
```

```python
model=RandomForestClassifier()
classify_model(model)
```

model=ExtraTreesClassifier()
classify_model(model)

```python
model=DecisionTreeClassifier()
classify_model(model)
```

```python
model=SVC()
classify_model(model)
```

```python
model=GradientBoostingClassifier()
classify_model(model)
```

```python
model=AdaBoostClassifier()
classify_model(model)
```
